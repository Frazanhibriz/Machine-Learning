```python
df.plot(kind="scatter", x="Cost of Living Value", y="Purchasing Power Value")
```

###Hasil Analisis

Korelasi positif pada scatter plot, ini menunjukan bahwa terdapat hubungan antara data cost of living dan purchasing power. Semakin tinggi biaya hidup, maka semakin tinggu pula daya belinya.

```python
sns.pairplot(df, diag_kind="kde", markers="o", corner=False)
plt.show()

print()

correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi')
plt.show()
```

###Hasil Analisis

Insight yang di dapat dari visualisasi data di atas adalah ada beberapa fitur data yang memiliki hubungan korelasi dengan data lainnya. Data Cost of Living dan Purchasing Power memiliki nilai koreasi yang paling tinggi dari yang lainnya, ini di tunjukan dari visualisasi heatmap yang memiliki nilai korelasi 0.60. Dan hubungan korelasi antara Data Cost of Living dan Purchasing Power yang di tunjukan dari scatter plot memiliki korelasi yang positif, ini artinya semakin tinggi biaya hidup, maka daya beli semakin tinggi di beberapa negara.
